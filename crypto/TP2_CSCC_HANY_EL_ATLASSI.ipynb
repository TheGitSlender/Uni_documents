```python
def lfsr(poly,seed,n):
    result=[]
    for i in range(n):
        result.append(seed[-1])
        new = 0 
        bascule = poly[:-1]
        for j in bascule:
            new = new ^ seed[j]
        seed = [new] + seed[:-1]
    return result
print(lfsr([0,2,3,4], [1,0,1,0,1], 10))
```

```python
test = "ENSAMCASABLANCA"
def word_to_bits(word):
    bits = []
    for letter in word:
        bits += [ int(i) for i in format(ord(letter)-65, '08b')]
    return bits
word_to_bits(test)
print(word_to_bits(test))
print(len(word_to_bits(test)))
```

```python
def chiffrer_LFSR(word,poly,seed):
    first_flux=lfsr(poly,seed,len(word)*8)
    word_b=word_to_bits(word)
    word_n=[]
    for i in range(len(word_b)):
        runner = word_b[i] ^ first_flux[i]
        word_n.append(runner)
    return word_n
word_n=chiffrer_LFSR(test,[0,2,3,4], [1,0,1,0,1])
print(word_n)
```

```python
def dechiffrer_LFSR(word_b,poly,seed):
    first_flux=lfsr(poly,seed,len(word_b))
    word_d=[]
    for i in range(len(word_b)):
        runner = word_b[i] ^ first_flux[i]
        word_d.append(runner)
    return word_d
print(dechiffrer_LFSR(word_n,[0,2,3,4], [1,0,1,0,1]))
```

```python
def per_lfsr(poly):
    seed = [1]+[0]*int(poly[-1]-1)
    ref=seed.copy()
    n=0
    while(True and n<2**len(seed)-1):
        b = poly[:-1]
        new = 0
        for j in b:
            new = new ^ seed[j]
        seed = [new] + seed[:-1]
        n=n+1
        if seed == ref:
            break
    return n
print(per_lfsr([0,2,3,4]))
```

```python
import random
import string
from Crypto.Cipher import Salsa20

TEXT_SIZE = 10 * 1024 * 1024  


print(f"Generating text of size {TEXT_SIZE/(1024*1024)} Mo...")

random_text = ''.join(random.choice(string.ascii_uppercase) for _ in range(TEXT_SIZE))
print(f"length of generated text: {len(random_text)}")
```

```python
def triviome(sequence,n):
    n=len(sequence)
    s=sequence.copy()
    result=[]
    for i in range(n):
        t1=s[65]^s[92]
        t2=s[161]^s[176]
        t3=s[242]^s[287]
        result.append((t1+t2+t3)%2)
        t1=t1^(s[90]&s[91])^s[170]
        t2=t2^(s[174]&s[175])^s[263]
        t3=t3^(s[285]&s[286])^s[68]
        Snew=[t3]+s[:93]+[t1]+s[93:175]+[t2]+s[175:287]
        s=Snew
    return result
```

```python
poly = [1,8, 11, 13, 14, 16]
seed = [random.randint(0, 1) for _ in range(16)]

sample_size = min(1000000, len(random_text)) 
sample_text = random_text[:sample_size]

keystream_bits = lfsr(poly, seed.copy(), sample_size * 8)

encrypted =chiffrer_LFSR(sample_text,poly,seed)
decrypted = dechiffrer_LFSR(encrypted,poly,seed)
```

```python
state = [0] * 288

key = [random.randint(0, 1) for _ in range(80)]
iv = [random.randint(0, 1) for _ in range(80)]

for i in range(80):
    state[i] = key[i]
for i in range(80):
    state[93 + i] = iv[i]
state[285] = state[286] = state[287] = 1

# Initialization
for i in range(4 * 288):
    t1 = state[65] ^ state[92]
    t2 = state[161] ^ state[176]
    t3 = state[242] ^ state[287]
    t1 = t1 ^ (state[90] & state[91]) ^ state[170]
    t2 = t2 ^ (state[174] & state[175]) ^ state[263]
    t3 = t3 ^ (state[285] & state[286]) ^ state[68]
    state = [t3] + state[:93] + [t1] + state[93:175] + [t2] + state[175:287]


trivium_state = state.copy()
keystream_bits = triviome(trivium_state, sample_size * 8)
sample_textBits = word_to_bits(sample_text)

encrypted = [0] * sample_size
for i in range(sample_size):
    byte = 0
    for j in range(8):
        byte |= (keystream_bits[(i * 8 + j)] << (7 - j))
    encrypted[i] = sample_textBits[i * 8 + j] ^ byte

decrypted = [0] * sample_size
for i in range(sample_size):
    byte = 0
    for j in range(8):
        byte |= (keystream_bits[(i * 8 + j) % (sample_size * 8)] << (7 - j))
    decrypted[i] = encrypted[i] ^ byte
```
